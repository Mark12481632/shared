```python
import seaborn as sns
import pandas as pd
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
lst_x = [x for x in range(1,101)]
lst_y = [x + rnd.random()*10 for x in lst_x]
```

```python
df = pd.DataFrame(list(zip(lst_x, lst_y)), columns = ['X', 'y'])
```

```python
sns.regplot(x='X', y='y', data=df)
plt.ylim(0,)
```

```python
sns.residplot(df['X'], df['y'])
```

```python
lr = LinearRegression()
XX = np.array(lst_x).reshape(100,1)
YY = np.array(lst_y).reshape(100,1)
lr.fit(XX,YY)
Yhat = lr.predict(XX)

ax1 = sns.distplot(df['y'], hist=False, color='r', label='Actual Values')
sns.distplot(Yhat, hist=False, color='b', label='Fitted values', ax=ax1)
```
